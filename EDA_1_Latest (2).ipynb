# Outlier Detection

```python
# histogram of Ozone
data_cleaned3['Ozone'].hist()
```

```python
#Box plot
plt.grid()
box = plt.boxplot(data_cleaned3["Ozone"],vert=False)
```

```python
# Extracting outliers
plt.grid()
box = plt.boxplot(data_cleaned3['Ozone'],vert=True)
```

```python
type(box)
```

```python
data_cleaned3["Ozone"].mean()
```

```python
# Printing outliers using standard deviation
for x in data_cleaned3["Ozone"]:
    mean = data_cleaned3["Ozone"].mean()
    std = data_cleaned3["Ozone"].std()
    if (x<(mean-3*std)) or (x > (mean+3*std)):
        print("This are the outliers ",x, end=' ')
```
